```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame of ATEs and CIs for all methods
methods = [
    "Linear Regression", "IPW",
    "T-Learner (RF)", "T-Learner (XGB)",
    "S-Learner (RF)", "S-Learner (XGB)",
    "Causal Forest"
]
ates = [0.4127, 0.4145, 0.4180, 0.4180, 0.4180, 0.4180, 0.4169]
lower_bounds = [0.3877, 0.3866, 0.4094, 0.4100, 0.4085, 0.4113, 0.2035]
upper_bounds = [0.4377, 0.4423, 0.4340, 0.4336, 0.4335, 0.4307, 0.6303]
```

```python
# Store in DataFrame
df = pd.DataFrame({
    "Method": methods,
    "ATE": ates,
    "CI Lower": lower_bounds,
    "CI Upper": upper_bounds
})
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.errorbar(df["Method"], df["ATE"], 
             yerr=[df["ATE"] - df["CI Lower"], df["CI Upper"] - df["ATE"]],
             fmt='o', capsize=5, color='black', ecolor='gray', linewidth=2)

plt.xticks(rotation=30, ha='right')
plt.ylabel("Estimated ATE")
plt.title("Comparison of ATE Estimates Across Methods")
plt.tight_layout()
plt.grid(True)
```
